```python
import pandas as pd
import torch
from sklearn.datasets import load_boston
from sklearn import preprocessing

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
dataset = load_boston()
```

```python
df = pd.DataFrame(dataset.data)
df.columns = dataset.feature_names
df["PRICES"] = dataset.target
df.head()
```

```python
TargetName = "PRICES"
FeaturesName = [
              "CRIM",
              'ZN',
              'INDUS',
              'CHAS',
              'NOX',
              'RM',
              'AGE',
              "DIS",
              'RAD',
              'TAX',
              'PTRATIO',
              'B',
              'LSTAT'
              ]
```

```python
X = df[FeaturesName]
Y = df[TargetName]
```

```python
sscaler = preprocessing.StandardScaler()
sscaler.fit(X)
X_std = sscaler.transform(X)
X_torch = torch.from_numpy(X_std.astype('float32'))
Y_torch = torch.from_numpy(Y.values.astype('float32'))
```

```python
class LinearReg(nn.Module):
        def __init__(self, input_shape):
            super().__init__()
            self.layer1 = nn.Linear(input_shape, 1)
        def forward(self, x):
            return self.layer1(x).squeeze(-1)

model = LinearReg(X_std.shape[1])
```

```python
#before training
#average pred
print(model(X_torch).mean())
#average observed
print(Y_torch.mean())
```

```python
optimizer = optim.SGD(model.parameters(), lr=0.005)
loss_function = nn.MSELoss()
```
